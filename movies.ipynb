```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('movies_metadata.csv')
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df=df[['title','overview','genres','tagline','vote_average','popularity']]
```

```python
df = df.dropna(subset=['title'])
```

```python
df['overview'] = df['overview'].fillna('')
```

```python
df.iloc[1]['genres']
```

```python
import ast
```

```python
df['genres'] = df['genres'].apply(lambda x: " ".join([i['name'] for i in ast.literal_eval(x)]))
```

```python
df['tagline'] = df['tagline'].fillna('')
```

```python
df.isnull().sum()
```

```python
df['tags'] = df['overview']+" "+df['genres']+" "+df['tagline']
```

```python
df
```

```python
import nltk 
from nltk.corpus import stopwords 
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z\s]','',text)
    words = text.split()
    
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    
    return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df = df.reset_index(drop=True)
```

```python
indices = pd.Series(df.index,df['title']).drop_duplicates()
```

```python
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=50000,stop_words='english',ngram_range=(1,2))
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title,n=10):
    if title not in indices:
        return ["Movie not found"]
    idx = indices[title]
    sim_scores = cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
    similar_idx = sim_scores.argsort()[::-1][1:n+1]
    
    return df['title'].iloc[similar_idx]
```

```python
recommend('Avatar')
```

```python
import pickle 

pickle.dump(tfidf_matrix,open('tf_idfmatrix.pkl','wb'))
pickle.dump(df,open('df.pkl','wb'))
pickle.dump(indices,open('indices.pkl','wb'))
```
